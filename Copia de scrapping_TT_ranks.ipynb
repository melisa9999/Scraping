```python
import requests
from bs4 import BeautifulSoup

URL = 'https://www.ittf.com/wp-content/uploads/2021/06/2021_23_SEN_WS.html'
page = requests.get(URL)  # se trae la información de la URL
soup = BeautifulSoup(page.content, 'html.parser') #se crea el objeto que contiene todo el código HTLM
#print(soup.prettify()) #permite imprimir el código en el formato correcto
```

```python
players= soup.tbody #accede al tag 'tbody'

lista_players = players.find_all(class_='rrow') # en 'tdoby' se accede a la clase con la lista de jugadoras

names= [lista_players[name].find_all('td')[1].text for name in range(len(lista_players))]

countries= [lista_players[country].find_all('td')[2].text for country in range(len(lista_players))]

scores =[lista_players[score].find_all('td')[3].text for score in range(len(lista_players))]

ranks= [position for position in range(1,len(lista_players)+1)]

```

```python
import pandas as pd
position_table = pd.DataFrame({"Rank position":ranks,"Names": names,"Country": countries,"Score": scores})
position_table
```

```python
frecuency_table=pd.crosstab(index=position_table['Country'], columns='count')
frecuency_table
```
